### 독립변수 개념 설명

##### 시총대비 거래량 (VolumePerMarketcap) 
 코인가격 * 코인수량 총액(시가총액) 대비 거래량. 
 
 시총대비 거래량이 상대적으로 높다는 것은 해당 시점에 순간적으로 거래가 많이 이루어졌다는 것을 뜻한다. 이는 매수자가 많은 것일 수도 있고 매도자가 많은 것일 수도 있다. 
 
 따라서 해당 시점의 방향은 장담 못하지만 최소한 움직임이 크다는 이야기는 될 수 있다.하지만 시총대비 거래량이 상대적으로 높은 상황이 지속적으로 유지가 된다면 다른 코인에 비해 해당 코인이 시장의 관심을 많이 받아 시장 참여자가 많다는 이야기가 될 수 있다. 이러한 경우 상승의 확률이 높아진다.


##### 롱숏비율 (LS_Ratio) 
 공매도 (미래의 가격을 예측하여 현재가격으로 매도 후 시세차익을 얻는 방식)와 공매수 (자금을 차입해 매수주문으로 상승장에서의 시세차익을 얻을 수 있다) 비율 

 이 비율은 상승, 하락 중 어디에 많은 사람들이 배팅하고 있는지를 보여준다. __시장심리를 대변한다__고 할 수 있다. 대부분, 사람들이 많은(비율이 높은) 쪽으로 시장심리가 쏠려있다고 볼 수 있으나 아래에 기술한 특정한 상황의 경우 반대 방향으로 큰 움직임이 나올 수 있다. 
 
 또한, 이 롱/숏 비율이 높으면 해당 방향으로 시장의 비이성적 움직임이 작용하고 있다고 보면 된다. 즉, 가격변동성이 높아진다.
  

스퀴징 : 스퀴징이란 강제청산가격에 도달하면 자동으로 시장가 반대매매가 나가면서 반대방향으로 가격이 밀려버리는 현상을 말한다. 예를 들어 공매도 즉, 숏이 스퀴징이 나는 것을 숏 스퀴징이라고 말하는데 공매도의 경우 매입 가격보다 가격이 2배가 상승할 경우 -100%의 손실을 보게 된다. 기존 예치금으로 -100% 이상의 손해를 볼 수 없기 때문에 -100%에 도달하는 순간 거래소 자체에서 자동적으로 해당 포지션을 청산하는 것이다. 공매도 상황에서의 청산은 매수를 뜻하고 이는 가격을 끌어올리는 요인이 된다. 특히, 해당 포지션에 동참한 사람이 많으면 많을수록 더 큰 상승을 보인다. 롱/숏 비율에서 숏이 압도적으로 많은 상황에서 상승을 하며 숏 비율이 줄어든다면 이는 숏스퀴징으로 볼 수 있다.


##### 구글 검색량 (total_search) 
 사람들의 코인에 대한 관심도를 볼 수 있으며 상승장에 많고 하락장에 적어진다.


##### 일론 머스크 트윗 (ntweets)
 투자자들의 심리에 영향을 미치는 요인. 
 
 특히, 2020-2021년에 가장 큰 영향을 미친 외부요인이다. 트윗 자체가 시장에 호재나 악재로 작용하는 경우가 많았으며 초기에는 대부분 호재로 작용하였으나 이후에는 악재로 작용하는 경우가 더 많아졌다. 확실한 것은 2020-2021년도에 머스크 트윗은 상승이든 하락이든 어떤 방향으로든지 변동성을 확대했을 것이라는 점이다.


##### 이동평균선 이격도 (disparity) 
 차트의 일정기간 내의 움직임이 평균선과 얼마나 멀어져 있는가에 대한 지표. 
 
 추세가 형성된 상황 (이격도가 양수 혹은 음수를 유지하는 상황)에서 그 추세의 강도를 나타내는 지표를 의미한다. 이는 시장 참여자들의 불안정한 투자심리를 더욱 악화시킬 수 있다. 이러한 경우 비이성적인 매도와 매수가 증가한다.


##### 김치프리미엄 (해외거래소와의 가격차이) (Kimchi)
 국내 거래소와 해외 거래소 간의 가격 차이. 
 
 국내 거래소는 개인의 거래가 많아 상대적으로 비이성적인 움직임이 많이 나타난다. 시장이 초 상승장이어서 비이성적 과열 상태가 나타날 때 김치프리미엄이 50%까지 발생하기도 하고 (해외 거래소에 비해 국내 가격이 50% 비싸지고) 하락장일 때는 이 반대 현상이 나타나기도 한다.
 
 즉, 김치프리미엄은 개인 투자자들의 투자 심리가 얼마나 비이성적 상태 혹은 공포에 달했는가를 보여주는 지표이다.


##### 모멘텀지표(Momentum) 
 기존 날짜의 종가 대비 현재 가격의 위치. 
 
 보통 추세(trend)를 나타내는데 추세의 개념은 기존 경제학에서 인정받지 못하다가 비교적 최근에 인정받게 된 개념이다. 
 
 가격의 상승이나 하락이 시장 참여자들의 불안정한 투자 심리 (FOMO 혹은 FUD)를 유발하여 이들이 비이성적인 투자를 하는 경우 추세가 형성된다.
 
 이러한 관점에서 모멘텀지표가 장기적으로 양수를 유지하는 경우 상승추세, 음수를 유지하는 경우 하락추세라고 볼 수 있으며, 이 추세가 어떻게 가격에 영향을 미치는지가 중요하다.


##### 해시 레이트(hash-rate) 
 비트코인의 채굴력을 의미한다. 채굴자가 많을수록 (연산처리량이 많을수록) 해시 레이트가 증가한다. 
 
 해시 레이트가 증가할수록 채굴의 난이도는 높아지기 때문에 가격 상승의 원인이 된다. 또한, 생산의 효율 및 비용 증가로 인한 공급자의 감소 및 신규 공급자의 유입 감소까지 예상할 수 있다.


##### Transaction count(채굴자들의 지갑간 이동)(BTC/Tx Cnt) 
 트랜젝션은 다수의 개인 투자자보다 소수의 초기 채굴자에 의해 변동되는 경우가 대다수이다. 이들은 압도적으로 많은 코인을 보유하고 있다. 
 
 그렇기 때문에 이들이 개인 지갑에 보유하고 있는 비트코인을 거래소로 보내며 트랜젝션이 오른다면 비트코인의 대량 매도로 인한 하락으로 볼 수 있고, 비트코인을 거래소에서 매수하여 개인 지갑으로 보내는 과정에서 트랜젝션이 증가한다면 비트코인을 매수 이후 홀딩한다는 뜻으로 볼 수 있으므로 가격상승을 예상할 수 있다.


##### 채굴자 수익(BTC/ Miner Revenue) 
 해시 당 채굴자들의 수익이 높아질수록 채굴량이 많아진다. 이는 채굴 난이도 또한 상승한다는 의미이며, 가격 상승으로 이어진다.

# 왜 회귀분석을 돌리는가?

### 회귀분석이란?
회귀분석(regression analysis)은 매개변수 모델을 이용하여 __통계적으로 변수들 사이의 관계를 추정__하는 분석방법이다. 

주로 독립변수(independent variable)가 종속변수(dependent variable)에 미치는 영향을 확인하고자 할 때 사용한다. 


### 회귀분석사용하는 이유
이 프로젝트에서 다양한 변수들이 비트코인의 가격에 얼마나 영향을 미치는지 분석하고 비트코인이 상용화폐를 대체할 수 있는지 알아보고자 한다. 

따라서 시총대비 거래량, 롱숏비율 등의 다양한 요인들이 비트코인 가격과 어떤 관계인지 추정하기 위해 회귀분석을 사용한다.



### OLS
최소제곱법 (Ordinary Least Squares)은 잔차 (실제값 – 예측값)의 제곱의 합이 최소화되도록 만드는 것이다. 추세선 (전체 데이터를 설명해주는 선)과 실제 데이터의 최소화되도록 그리는 방법이다.


### OLS step
1. 데이터 준비하기
     - X에는 독립변수, Y에는 종속변수로 구분하기
2. 모델 구조 만들기
3. 모델 학습 시키기
4. 모델 요약하기

# 비트코인 선형 회귀 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
# 종속변수 불러오기
btc_usd = pd.read_csv('C:/Users/apple/datascience_bitcoin/btc-usd.csv')

# 종속변수 Date(날짜)와 Close(종가 = 가격) 빼고 제거
btc_usd = btc_usd.loc[:, ['Date', 'Close']]
```

```python
# 독립변수 불러오기
volume = pd.read_csv('C:/Users/apple/datascience_bitcoin/volume.csv')
google_search = pd.read_csv('C:/Users/apple/datascience_bitcoin/google-search.csv')
musk_tweet = pd.read_csv('C:/Users/apple/datascience_bitcoin/musk-tweets.csv')
ls_ratio = pd.read_csv('C:/Users/apple/datascience_bitcoin/ls-ratio.csv')
disparity = pd.read_csv('C:/Users/apple/datascience_bitcoin/disparity.csv')
kimchi = pd.read_csv('C:/Users/apple/datascience_bitcoin/kimchi.csv').loc[:, ['Date', 'Kimchi']]
momentum = pd.read_csv('C:/Users/apple/datascience_bitcoin/momentum.csv').loc[:, ['Date', 'Momentum']]
hash_rate = pd.read_csv('C:/Users/apple/datascience_bitcoin/hash-rate.csv')
transaction_revenue = pd.read_csv('C:/Users/apple/datascience_bitcoin/transaction-revenue.csv')

# 독립변수 병합
total = pd.merge(volume, google_search, on = 'Date').fillna(0)
total = pd.merge(total, musk_tweet, on = 'Date').fillna(0)
total = pd.merge(total, ls_ratio, on = 'Date').fillna(0)
total = pd.merge(total, disparity, on = 'Date').fillna(0)
total = pd.merge(total, kimchi, on = 'Date').fillna(0)
total = pd.merge(total, momentum, on = 'Date').fillna(0)
total = pd.merge(total, hash_rate, on = 'Date').fillna(0)
total = pd.merge(total, transaction_revenue, on = 'Date').fillna(0)
```

```python
btc_usd
```

```python
total
```

```python
X = total.drop(['Date'], axis=1)
Y = btc_usd.drop(['Date'], axis=1)
```

```python
# intercept 추가
X = sm.add_constant(X, has_constant='add')
```

```python
# 학습 데이터, 테스트 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=1234)
```

```python
# 회귀 모델 생성
lm = sm.OLS(Y_train, X_train)
lm_train = lm.fit()
```

```python
lm_train.summary()
```

```python
Y_predict = lm_train.predict(X_test)
print(Y_predict)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(Y_test, Y_predict)
plt.xlabel('Y', size=16)
plt.ylabel('$\hat{Y}$', size=16)
plt.show()
```

```python
# MSE(mean squared error), R-squared (모형 성능 확인)
mae = mean_absolute_error(Y_test, Y_predict)
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
r_square = r2_score(Y_test, Y_predict)
print('MAE : {0:.3f}'.format(mae))
print('MSE : {0:.3f}'.format(mse))
print('RMSE : {0:.3f}'.format(rmse))
print('R-square : {0:.3f}'.format(r_square))
```

## 회귀분석 결과 해석
그래프에서 확인했던 것처럼 시총대비거래량이 비트코인 가격과 반비례관계이다.<br>
또한,생산과 공급쪽의 변수에 가격이 큰 영향을 받는 것을 확인할 수 있다. 
<br> 채굴자들의 생산, 채굴 수익, 매도량이 주식에서의 세력과도 같은 역할을 한다고 볼 수 있을 것이다. 


일론 머스크 트윗과 모멘텀 지표의 p-value 값이 0.05 이상으로 학술적으로는 유의미하지 않은 결과가 나왔다. <br>
모멘텀 지표 자체는 0.059로 근사적으로(marginally) 유의미했다. 


일론 머스크 트윗 변수를 제외한 후 다시 회귀분석을 돌려본 결과, 모든 변수들이 유효한 수치를 보여줬으며, <br>
R-squared는 0.969로 1에 가깝기 때문에 예측 정확도가 매우 높다고 볼 수 있다. <br>
마찬가지로 맷플롯 시각화 결과도 한 두개를 제외하고 양호한 결과가 나온 모습이다. <br>
결론적으로 데이터를 잘 설명하는 정확도가 높은 모델이라고 볼 수 있다.

## 이더리움 선형 회귀 분석 (비트코인 가격 + 생산과 공급 측면 관점 변수 3개)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
# 새로운 종속변수(이더리움 종가) 불러오기
eth = pd.read_csv('C:/Users/apple/datascience_bitcoin/ethereum.csv')

eth
```

```python
# 새로운 독립변수 병합
total2 = pd.merge(btc_usd, volume, on = 'Date').fillna(0)
total2 = pd.merge(total2, ls_ratio, on = 'Date').fillna(0)
total2 = pd.merge(total2, kimchi, on = 'Date').fillna(0)
total2 = pd.merge(total2, hash_rate, on = 'Date').fillna(0)
total2 = pd.merge(total2, transaction_revenue, on = 'Date').fillna(0)

total2
```

```python
X3 = total2.drop(['Date', 'VolumePerMarketcap', 'ls-ratio', 'Kimchi'], axis=1)
Y3 = eth.drop(['Date'], axis=1)
```

```python
# intercept 추가
X3 = sm.add_constant(X3, has_constant='add')
```

```python
# 학습 데이터, 테스트 데이터 분리
X3_train, X3_test, Y3_train, Y3_test = train_test_split(X3, Y3, train_size=0.7, test_size=0.3, random_state=1234)
```

```python
# 회귀 모델 생성
lm3 = sm.OLS(Y3_train, X3_train)
lm3_train = lm3.fit()
```

```python
lm3_train.summary()
```

```python
Y3_predict = lm3_train.predict(X3_test)
print(Y3_predict)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(Y3_test, Y3_predict)
plt.xlabel('Y3', size=16)
plt.ylabel('$\hat{Y3}$', size=16)
plt.show()
```

```python
# MSE(mean squared error), R-squared (모형 성능 확인)
mae = mean_absolute_error(Y3_test, Y3_predict)
mse = mean_squared_error(Y3_test, Y3_predict)
rmse = np.sqrt(mse)
r_square = r2_score(Y3_test, Y3_predict)
print('MAE : {0:.3f}'.format(mae))
print('MSE : {0:.3f}'.format(mse))
print('RMSE : {0:.3f}'.format(rmse))
print('R-square : {0:.3f}'.format(r_square))
```
